```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = sns.load_dataset('iris')
```

```python
df1.head(1)
```

## PanelGrid

```python
graph1 = sns.PairGrid(df1)
graph1.map(plt.bar)
```

```python
graph1 = sns.PairGrid(df1)
graph1.map(plt.scatter)
```

```python
graph1 = sns.PairGrid(df1)
graph1.map_diag(sns.distplot)
```

```python
graph1 = sns.PairGrid(df1)
graph1.map_upper(plt.scatter)
```

```python
graph1 = sns.PairGrid(df1)
graph1.map_upper(plt.scatter)
graph1.map_diag(sns.kdeplot)
graph1.map_lower(plt.scatter)
```

```python
graph1 = sns.PairGrid(df1)

graph1.map_upper(plt.scatter)
graph1.map_diag(sns.kdeplot)
graph1.map_lower(sns.kdeplot)
```
