```python
import seaborn as sns
```

```python
ds = sns.load_dataset('tips')
```

```python
ds.head()
```

```python
ds.columns
```

## Histogram

```python
sns.distplot(ds['total_bill'])
```

```python
sns.distplot(ds['tip'])
```

```python
sns.jointplot(x='total_bill', y='tip', data=ds)
```

## Lineal Reg

```python
sns.jointplot(x='total_bill', y='tip', data=ds, kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip', data=ds, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=ds, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='size', data=ds, color='red')
```

```python
sns.pairplot(ds)
```

```python
sns.pairplot(ds, hue='sex')
```
