```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head(2)
```

```python
sns.lmplot(x='total_bill', y='tip', data=df)
```

```python
sns.lmplot(x='total_bill', y='tip', data=df, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=df, hue='sex', markers=['x', 'o'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=df, hue='sex', markers=['x', 'o'], scatter_kws={'s':30})
```

```python
sns.lmplot(x='total_bill', y='tip', data=df, col='day', row='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=df, col='day', hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=df, col='day', hue='smoker')
```

```python
sns.lmplot(x='total_bill', y='tip', data=df, aspect=0.9, size=7)
```
