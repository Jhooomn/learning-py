```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(2, 50, 10)
x
```

```python
y = x ** 2
y
```

```python
figure, graph = plt.subplots(nrows=2, ncols=2)

graph[0][0].plot(x,y, 'red')
graph[0][1].plot(y,x, 'green')
graph[1][0].plot(x,y, 'blue')
graph[1][1].plot(y,x, 'black')


graph[0][0].set_title('Red')
graph[0][1].set_title('Green')
graph[1][0].set_title('Blue')
graph[1][1].set_title('Black')
```

```python
figure = plt.figure(figsize=(4,2))
graph = figure.add_axes([0.1, 0.1, 0.9, 0.9])
graph.plot(x,y, 'red')
```

```python
figure = plt.figure(figsize=(4,2))
graph = figure.add_axes([0.1, 0.1, 0.9, 0.9])
graph.plot(x,y**2, 'blue')
```
