```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(2,15, 20)
x
```

```python
y = x**2
y
```

```python
plt.plot(x, y, 'red')
plt.title('My first plot!')
plt.xlabel('This is X')
plt.ylabel('This is Y')
plt.legend('Relation')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'green')
plt.title('Plot #1')

plt.subplot(1,2,2)
plt.plot(y,x,'blue')
plt.title('Plot #2')
plt.show()
```
