```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2, 20)
y = x**2
```

```python
figure = plt.figure()
graph = figure.add_axes([0,0,1,1])
graph.plot(x,y, marker='o', markersize=5, linewidth=1)
```

```python
figure = plt.figure(figsize=(10,6))
graph = figure.add_axes([0,0,1,1])
graph.plot(x,y, color='yellow', linewidth=1, linestyle='--', marker='o', markersize=5, alpha=1, label='X')
graph.plot(x**2, y, color = 'blue', linewidth=1, linestyle='--', marker='o', markersize=5, alpha=1, label='X**2')
graph.plot(x**3, y, color = 'red', linewidth=1, linestyle='--', marker='o', markersize=5, alpha=1, label='X**3')

graph.legend(loc=0)
```
