```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(2, 15, 10)
y = x**2
```

```python
figure = plt.figure()
graph = figure.add_axes([0.1, 0.1, 0.9, 0.9])

graph.plot(x**2, label='X**2')
graph.plot(x**3, label='X**3')

graph.legend(loc=0)
```

```python
figure = plt.figure()
graph = figure.add_axes([0.1,0.1,0.9,0.9])

graph.plot(x,y,label='X')
graph.plot(x**2,y,label='X**2')
graph.plot(x**3,y,label='X**3')
graph.plot(x**4,y,label='X**4')

graph.legend(loc=0)
```
