```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10, 5)
y = x**2
```

```python
figure = plt.figure()
graph = figure.add_axes([0,0,1,1])
graph.plot(x,y, color='red', linewidth=5, alpha=0.4, linestyle='-.')
```

```python
figure = plt.figure()
graph = figure.add_axes([0,0,1,1])

graph.plot(x**2,y, color='blue', linewidth=7, alpha=0.7, linestyle='--')
```

```python
figure = plt.figure(figsize=(5,5))
graph= figure.add_axes([0,0,1,1])
graph.plot(x,y, color='red', linewidth=5, linestyle=':', alpha=0.6, label='X')
graph.plot(x**2, y, color='green', linewidth=5, linestyle='-.', alpha=0.7, label='X**2')
graph.plot(x**3,y, color='blue', linewidth=6, linestyle='--', alpha=0.8, label='X**3' )
graph.legend()
```
