```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(2,15, 20)
x
```

```python
y = x**2
y
```

```python
figure, graph = plt.subplots(nrows=2, ncols=2)

graph[0][0].plot(x,y, 'y')
graph[0][1].plot(x,y, 'b')
graph[1][0].plot(x,y, 'r')
graph[1][1].plot(x,y, 'g')


graph[0][0].set_title('Yellow')
graph[0][1].set_title('blue')
graph[1][0].set_title('Red')
graph[0][1].set_title('Green')
```
